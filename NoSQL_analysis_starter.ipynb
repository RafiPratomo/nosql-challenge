## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
print(db.list_collection_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.
* Convert the result to a Pandas DataFrame, print the number of rows in the DataFrame, and display the first 10 rows.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
hygiene_20_count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {hygiene_20_count}")
# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_results = establishments.find(query)
hygiene_20_df = pd.DataFrame(hygiene_20_results)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(hygiene_20_df)}")
# Display the first 10 rows of the DataFrame
print(hygiene_20_df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": "City of London Corporation", "RatingValue": {"$gte": 4}}
# Use count_documents to display the number of documents in the result
london_high_rating_count = establishments.count_documents(query)
print(f"Number of establishments in London with RatingValue >= 4: {london_high_rating_count}")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
london_high_rating_results = establishments.find(query)
london_high_rating_df = pd.DataFrame(london_high_rating_results)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(london_high_rating_df)}")
# Display the first 10 rows of the DataFrame
print(london_high_rating_df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.4860
longitude = 0.0060

query = {
    "RatingValue": 5,
    "$expr": {
        "$and": [
            {"$gte": [{"$toDouble": "$geocode.latitude"}, latitude - degree_search]},
            {"$lte": [{"$toDouble": "$geocode.latitude"}, latitude + degree_search]},
            {"$gte": [{"$toDouble": "$geocode.longitude"}, longitude - degree_search]},
            {"$lte": [{"$toDouble": "$geocode.longitude"}, longitude + degree_search]}
        ]
    }
}
sort = [("scores.Hygiene", 1)]
limit = 5

# Print the results
top_5_results = list(establishments.find(query).sort(sort).limit(limit))
for result in top_5_results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
top_5_results_df = pd.DataFrame(top_5_results)
print(top_5_results_df)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Print the number of documents in the result
hygiene_0_results = list(establishments.aggregate(pipeline))
print(f"Number of Local Authorities with establishments having a hygiene score of 0: {len(hygiene_0_results)}")

# Print the first 10 results
for result in hygiene_0_results[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
hygiene_0_df = pd.DataFrame(hygiene_0_results)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(hygiene_0_df)}")
# Display the first 10 rows of the DataFrame
print(hygiene_0_df.head(10))
```
